```python
import pandas as pd
import numpy as np
import requests

from bs4 import BeautifulSoup
#importing the wikipedia source 
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

#converting into a dataframe
df = pd.read_html(website_url)[0]
df.to_csv('beautifulsoup_pandas.csv',header=0,index=True)
print (df)
```

```python
s1 = pd.Series(name='Neighbourhood')
mydf = pd.concat([df, s1], axis=1)    #concatenate the series to the df
mydf
```

```python
df.dropna(how='all')  #drop the column with NaN
```

```python
df.replace('Not assigned',np.NaN, inplace=True)
df
```

```python
#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
df.Neighbourhood.fillna(df.Borough, inplace=True)
df[df['Borough'].notnull()]
df
```

```python
df[df['Borough'].notnull()].reset_index() 
```

```python
df = df.astype(str) #convert all columns to dtype-string
```

```python
mydf = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index(name='Neighbourhood')
mydf
```

```python
mydf.replace('nan',np.NaN, inplace=True)  #replace nan with NaN to use .dropna method
mydf
```

```python
mydf.dropna(subset=['Borough','Neighbourhood'], how='all', inplace = True)
mydf
```
